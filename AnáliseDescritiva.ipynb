# Análise Descritiva

É menos complexo de análise de dados, que consiste em descrever e resumir os dados para entender o que aconteceu.

* Algumas meddida: médias, medianas, contagens e percentuais.


```python
import pandas as pd
df = pd.read_csv('base_dados_vendas_tratada.csv', delimiter=',', parse_dates=['dt_venda', 'dt_entrega'], index_col='dt_venda')
```

```python
df.head(2)
```

```python
df.info()
```

```python
#Verificando Missing Values
df.isnull().sum()
```

```python
#Verificando Valores de Registros Duplicados
df.duplicated().sum()
```

```python
#Verificando as primeiras estatíticas da base
df.describe().round(1)
```

```python
df.index
```

```python
df.describe()[ ['preco_custo','valor_unitario','valor_final'] ].round(1)
```

```python
#Contando Colunas - COUNT()
df.count()
df.dt_entrega.count()
```

```python
#Somando Valores de Linhas ou colunas
df['quantidade'].sum()

colunas = ['preco_custo','valor_unitario','valor_final']
df[colunas].sum()
```

```python
#Otendo a média
df[colunas].mean()
```

```python
#Maior Valor
df[colunas].max()
```

```python
#Menor Valor
df[colunas].min()
```

```python
#Média
df[colunas].mean()
```

```python
#Mediana
df[colunas].median()
```

```python
#Moda
df[colunas].mode()
```

```python
df['descricao_produto'].value_counts()
```

#Filtrando Dados do DataFrame

```python
#Testes Lógicos
#Exemplo com uma condição
filtro_vf = df['valor_final'] < 1000
df[filtro_vf]
```

```python
#Exemplo com mútiplas condições
filtro_vf = df['valor_final'] < 1000
filtro_loja = df['nome_loja'] == 'Filial SP'
df[filtro_vf & filtro_loja].head(2)
```

```python
#Filtrando registros do DataFrame com o método query
df.query("valor_final < 1000 and nome_loja == 'Filial SP'").head(2)
```

```python
df.nome_loja.unique()
```

```python
df.query("valor_final < 1000 and nome_loja == 'Loja Matriz'").head(2)
```

```python
df['descricao_produto'].unique()
```

```python
df.query("descricao_produto == 'Fritadeira elétrica' or nome_loja == 'Loja Matriz'").head(2)
```

```python
#Utilizando o index para filtrar o DataFrame
filtro_ano = df.index.year == 2020
filtro_mes = df.index.month == 2
filtro_matriz = df['nome_loja']  == 'Loja Matriz'
df[filtro_ano & filtro_matriz & filtro_mes].head(2)
```

```python
df.reset_index().head(2)
```

```python
#Agrupando por mais de uma coluna
df.groupby(['nome_funcionario', 'descricao_produto'])[ ['valor_final','preco_custo','quantidade']].sum()
```
