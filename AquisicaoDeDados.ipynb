```python
import pandas as pd
```

```python
df_ficticias = pd.read_csv('vendas_ficticias.csv')
```

```python
df_ficticias
```

```python
df_ficticias.info()
```

```python
df_ficticias.columns
```

```python
colunas_fict = ['id', 'produto', 'categoria', 'preco', 'quantidade', 'data_venda',
       'vendedor']
```

```python
df_ficticias.columns = colunas_fict
```

```python
df_ficticias.columns
```

```python
#Coluna preco está como str, precisa mudar para flaot
df_ficticias['preco']
```

```python
df_ficticias['preco'] = df_ficticias['preco'].str.replace(',','.')
```

```python
df_ficticias['preco'] = df_ficticias['preco'].str.replace("'", '')
```

```python
print(df_ficticias['preco'])
```

```python
print(df_ficticias)
```

```python
df_ficticias['preco'] = df_ficticias['preco'].astype({'preco':'float64'})
```

```python
df_ficticias['id'].duplicated().sum()
```

```python
df_ficticias['data_venda'] = pd.to_datetime(df_ficticias['data_venda'], format='%Y-%m-%d')
```

```python
df_ficticias.info()
```

```python
df_ficticias['valor_total'] = df_ficticias['quantidade'] * df_ficticias['preco']
```

```python
df_ficticias
```

```python
df_ficticias['produto'].unique()
```

```python
df_ficticias['produto'].fillna('Produto nao Identificado', inplace=True)
```

```python
df_ficticias['produto'].unique()
```

```python
df_ficticias['categoria'].fillna('Categoria nao Identificada', inplace=True)
```

```python
df_ficticias['categoria'].unique()
```

```python
df_ficticias['vendedor'].unique()
```

```python
df_ficticias['vendedor'].fillna('Vendedor nao Identificado', inplace=True)
```

```python
df_ficticias['vendedor'].unique()
```

```python
#Preenchendo Data Nula com data acima mais próxima
df_ficticias['data_venda'].ffill()
```

```python
df_ficticias['data_venda'].isnull().sum()
```

```python
#Excluir Duplicadas
df_ficticias.drop_duplicates(inplace=True)
```

```python
#Excluir coluna do Data Frame
#df_ficticias.drop(columns=['id'], inplace=True)
```

```python
#Ordenando o DataFrame de acordo com uma coluna específica
df_ficticias.sort_values('produto', inplace=True)
```
