```python
df = pd.DataFrame({'variavel_a':['1,1','2,2','3.3'],
                   'variavel_b':['4','5','6'],
                   'variavel_c':['10/10/2021','11/12/2020','15/09/2019']})
df
```

```python
df.info()
```

```python
df['variavel_a'] = df['variavel_a'].str.replace("'"," ")
```

```python
df['variavel_a'] = df['variavel_a'].str.replace(",",".")
```

```python
df['variavel_a'] = df['variavel_a'].astype({'variavel_a':'float64'})
```

```python
df['variavel_b'] = df['variavel_b'].astype({'variavel_b':'int'})
```

```python
df['variavel_c'] = pd.to_datetime(df['variavel_c'], format='%d/%m/%Y')
```

```python
df.info()
```
