```python
#Pacotes de Manipulação de Dados
import numpy as np
import pandas as pd

#Pacotes Gráficos
import matplotlib.pyplot as plt
import seaborn as sns

#Pacotes de Modelagem
import statsmodels.api as sm
```

```python
df = pd.read_csv('Base_Notas_v1.csv', sep=',', index_col='id')
df.head()
```

```python
#Gráfico de Dispersão
sns.scatterplot(x=df['horas_estudo'], y=df['nota_exame'])
```

```python
df.corr()
```

```python
#Gráfico HeatMap
sns.heatmap(df.corr(), cmap='RdBu_r', annot=True, vmin=-1,vmax=1)
```

```python
y = df['nota_exame']

df['intercepto'] = 1
x = df[['intercepto','horas_estudo']]
```

```python
#Ajuste do modelo
modelo = sm.OLS(y,x)
resultado = modelo.fit()
```

```python
#Resultado do ajuste do modelo
print(resultado.summary())
```

# 1. R² (Coeficiente de Determinação)

## O R² mede a proporção da variabilidade nos dados que é explicada pelo modelo. Quanto mais próximo de 1, melhor o modelo se ajusta aos dados.

* Base 1: R² = 0.5201

* Base 2: R² = 0.7585

* Base 3: R² = 0.7932

## ***Base 3 tem o maior R², o que significa que ela é a que melhor explica a variabilidade dos dados.***

# 2. R² Ajustado

## O R² ajustado leva em consideração o número de variáveis independentes no modelo e penaliza a adição de variáveis que não contribuem significativamente para o modelo. Ele é útil para comparar modelos com diferentes números de variáveis.

 R² Ajustado

*   Base 1: R² Ajustado = 0.5030
*   Base 2: R² Ajustado = 0.7406
*   Base 3: R² Ajustado = 0.7693

## **Base 3 tem o maior R² ajustado, indicando que ela não só explica mais variabilidade dos dados como também é mais eficiente, já que penaliza melhor as variáveis desnecessárias.**

# 3. AIC (Akaike Information Criterion)

## O AIC é um critério para comparar modelos, penalizando pela complexidade (número de parâmetros). Menores valores de AIC indicam um melhor ajuste, com menor complexidade.

* Base 1: AIC = 197.1051
* Base 2: AIC = 178.5007
* Base 3: AIC = 175.8525

##**Base 3 tem o menor AIC, o que sugere que, embora tenha mais parâmetros do que a Base 1 e 2, ela oferece um bom ajuste sem ser excessivamente complexa.**

# 4. BIC (Bayesian Information Criterion)

## O BIC também penaliza a complexidade do modelo, com uma penalização mais forte do que o AIC para modelos com muitas variáveis. Valores mais baixos de BIC indicam um modelo mais eficiente.

* Base 1: BIC = 199.9075
* Base 2: BIC = 182.7043
* Base 3: BIC = 181.4573

## ***Base 3 tem o menor BIC, o que reforça a ideia de que ela é o modelo com melhor trade-off entre ajuste e complexidade***
